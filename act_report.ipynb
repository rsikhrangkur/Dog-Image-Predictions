I love dogs. You *probably* love dogs. And we're far from the only ones.

There are tens of millions of people across the globe who are fond of man's best friend, and for dog lovers everywhere there's little better content to find than following [WeRateDogs](https://twitter.com/dog_rates) on Twitter.

Compelled to see what attracted users to their account and what makes their tweets a hit, I wrangled the data found in their twitter archive using Python and other libraries available to me.

### What kind of information is there?

After I cleaned what I could of the archive, we have nearly 2,000 tweets complete with text, pictures, favorite counts, retweets.

In the data there's a good amount of tweets where the dog's name is mentioned, and they even have a dog stage. These dog stages are "doggo", "floofer", "pupper" and "puppo".

Also included in the data were three algorithms which all used the pictures in these tweets to predict if the subject of the picture was a dog, what breed of dog (or what kind of object if it wasn't a dog), and the confidence of the algorithm being correct.

That is a **lot** of information to process. But from this archive, there were several insights I could make such as:

- Labrador Retriever and Golden Retriever were among the top 3 most common predictions of dog breed for all three algorithms.
    - Though while they're of the most commonly predicited breeds, the average confidence of these predictions isn't all that high.<br><br>
    
- There are a total of 1,350 tweets that have a name for their dog. That's a lot of names!
    - 913 of these names are entirely unique!<br><br>

- The "doggo" stage has the highest average retweet count, and the "puppo" stage has the highest average favorite count.
    - One dog in the archive is in both of these stages, and has a higher favorite and retweet count than these averages!
